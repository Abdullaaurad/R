```r
# Sample data
values <- c(30, 60, 45, 80, 50)
categories <- c("A", "B", "C", "D", "E")

# Define colors
colors <- c("red", "blue", "green", "yellow", "purple")

par(bg="white")
# Create the pie chart
pie(values, 
    labels = categories,   # Labels for slices
    col = colors,          # Colors for slices
    main = "Pie Chart with Custom Radius",
    clockwise = TRUE,      # Make it clockwise
    init.angle = 90,       # Start at 90 degrees (top)
    radius = 1)          # Increase the radius (default is 1)
```

```r
# Sample data
values <- c(30, 60, 45, 80, 50)
categories <- c("A", "B", "C", "D", "E")

par(bg="white")
# Create the barplot
barplot(values, 
        names.arg = categories,     # X-axis labels
        col = "skyblue",          # Bar colors
        main = "Simple Bar Plot",   # Main title
        xlab = "Categories",        # X-axis label
        ylab = "Values"             # Y-axis label
)
```

```r
# Sample numeric data
values <- c(10, 15, 20, 22, 25, 27, 30, 35, 37, 40, 42, 45, 47, 50, 55)

# Generate a stem-and-leaf plot
stem(values)
```

```r
# Sample data
set.seed(123)  # Ensures reproducibility
values <- rnorm(100, mean = 50, sd = 10)  # Generate 100 random numbers
par(bg="white")
# Create the histogram
hist(values,
     main = "Customized Histogram",  # Title of the histogram
     xlab = "Values",                 # X-axis label
     xlim = c(20, 80),                # X-axis limits
     ylim = c(0, 20),                 # Y-axis limits
     breaks = 10,                     # Number of bins
     col = "skyblue",                 # Fill color of bars
     border = "red")                 # Border color of bars
```

```r
# Sample data
set.seed(123)  # Ensures reproducibility
data <- list(
  Group_A = rnorm(50, mean = 60, sd = 10),
  Group_B = rnorm(50, mean = 70, sd = 15),
  Group_C = rnorm(50, mean = 80, sd = 20)
)

par(bg="white")
# Create the boxplot
boxplot(data, 
        notch = TRUE,                 # Adds notches to show median confidence
        varwidth = TRUE,              # Adjusts box width based on sample size
        names = c("A", "B", "C"),     # Custom group names
        main = "Customized Boxplot",  # Title of the boxplot
        col = c("red", "blue", "green"))  # Colors for the boxes
```

```r
set.seed(123)  # Ensures reproducibility
x <- rnorm(50, mean = 50, sd = 10)  # Random x values
y <- rnorm(50, mean = 50, sd = 10)  # Random y values

par(bg = 'white')

# Create scatter plot
plot(x, y, 
     main = "Basic Scatter Plot",  # Title
     xlab = "X Axis Label",        # X-axis label
     ylab = "Y Axis Label",        # Y-axis label
     xlim = c(20, 80),             # X-axis limits
     ylim = c(20, 80),             # Y-axis limits
     col = "blue",                 # Point color
     pch = 16)                     # Point shape (solid circle)  

# Add vertical line at x = 40
abline(v = 40, col = "red", lwd = 2, lty = 2)  # Dashed red line
```
